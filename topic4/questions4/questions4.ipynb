1. Назвать и охарактеризовать этапы анализа данных в схеме «разделение – применение – объединение»

**Ответ**: Схема «разделение-применение-объединение» (split-apply-combine) включает три этапа анализа данных:

- **Разделение.** Набор данных разбивается на более мелкие фрагменты, основываясь на определённых критериях.
- **Применение.** Каждый фрагмент обрабатывается независимо. Для каждой группы можно выполнить одну или несколько операций:
  - *Агрегация*: вычисляет итоговую статистику, например, групповые средние или количество элементов в каждой группе.
  - *Преобразование*: выполнение вычислений по группам или элементам.
  - *Фильтрация*: удаление целых групп данных на основе вычислений по группам.
- **Объединение.** Все результаты объединяются обратно и представляют собой единое целое. Этот заключительный этап выполняется библиотекой pandas автоматически — она собирает результаты, полученные на этапе применения, в единое целое.

В библиотеке pandas разделение выполняется с помощью метода `.groupby()` объекта Series или объекта DataFrame. В этот метод передаётся одна или несколько индексных меток и/или имя столбца, в результате происходит группировка данных на основе соответствующих значений.

2. Группировка по значениям отдельного столбца. Приведите примеры.

**Ответ**: Данные датчиков состоят из трёх категориальных переменных (sensor, interval и axis) и одной количественной переменной (reading). Их можно сгруппировать с помощью любой категориальной переменной, передав имя переменной в метод `.groupby()`.

В результате применения метода `.groupby()` объекта DataFrame получаем подкласс объекта GroupBy, либо DataFrameGroupBy для датафрейма, либо SeriesGroupBy для серии. Этот объект представляет собой промежуточное описание группировки.

- Свойство `.ngroups` извлекает количество групп, которые будут сформированы в итоге.
- Свойство `.groups` возвращает питоновский словарь, ключи которого представляют имена каждой группы, а значения — массив индексных меток, принадлежащих каждой соответствующей группе.

Следующий программный код группирует данные датчиков по значениям столбца sensor:

```python
# загружаем данные датчиков
sensor_data = pd.read_csv('sensor.csv', sep=';')
print(sensor_data.iloc[:5])

# группировка этих данных по столбцу/переменной sensor
# возвращает объект DataFrameGroupBy
grouped_by_sensor = sensor_data.groupby('sensor')
print(grouped_by_sensor)

# получаем информацию о количестве групп, которые будут созданы
print(grouped_by_sensor.ngroups)

# что представляют из себя найденные группы?
print(grouped_by_sensor.groups)
```

3. Просмотр результатов группировки. Приведите примеры.

**Ответ**: Сгруппированную переменную можно рассматривать как набор поимённых групп и использовать для просмотра содержимого групп.

- Для каждого уникального значения столбца sensor создаётся отдельная группа; названием каждой группы становится соответствующее значение столбца sensor.
- Метод `.size()` возвращает сводку о размере каждой группы.
- Метод `.count()` возвращает количество элементов в каждом столбце каждой группы.
- С помощью метода `.get_group()` можно извлечь любую конкретную группу.
- Для вывода указанного количества элементов в каждой группе можно использовать методы `.head()` и `.tail()`.
- Метод `.nth()` возвращает n-ную строку в каждой группе.
- Для вывода описательных статистик по каждой группе можно использовать метод `.describe()`.
- Группы отсортированы по названиям в порядке возрастания. Если вы хотите предотвратить сортировку при выполнении группировки, воспользуйтесь параметром `sort=False`.

```python
sensor_data = pd.read_csv('sensor.csv', sep=';')
grouped_by_sensor = sensor_data.groupby('sensor')

# вспомогательная функция, печатающая содержимое групп
def print_groups(group_object):
    for name, group in group_object:
        print(name)
        print(group[:5])

# смотрим содержимое созданных групп
print_groups(grouped_by_sensor)

# получаем информацию о количестве элементов в каждой группе
print(grouped_by_sensor.size())

# получаем информацию о количестве элементов в каждом столбце каждой группы
print(grouped_by_sensor.count())

# получаем данные конкретной группы
print(grouped_by_sensor.get_group('accel')[:5])
```

4. Группировка по нескольким столбцам. Приведите примеры.

**Ответ**: Группировку можно выполнить по нескольким столбцам, передав список имён столбцов в метод `.groupby()`. Поскольку было указано несколько столбцов, название каждой группы теперь является кортежем, представляющим уникальную комбинацию значений столбцов.

Следующий программный код группирует данные по столбцам sensor и axis:

```python
sensor_data = pd.read_csv('sensor.csv', sep=';')

def print_groups(group_object):
    for name, group in group_object:
        print(name)
        print(group[:5])

# группируем по значениям столбцов sensor и axis
mcg = sensor_data.groupby(['sensor', 'axis'])
print(print_groups(mcg))
```

5. Группировка по уровням индекса. Приведите примеры.

**Ответ**: Группировку можно выполнить, используя вместо столбцов значения индекса. Данные датчиков хорошо подходят для иерархического индекса, который можно использовать для иллюстрации такого способа.

Чтобы создать иерархический индекс, уровнями которого будут столбцы sensor и axis, нужно использовать метод `.set_index()`. Затем можно выполнить группировку по различным уровням иерархического индекса с помощью параметра `level`. Если у каждого уровня MultiIndex есть имена, то их можно использовать вместо целочисленных значений.

Следующие примеры демонстрируют группировку по одному уровню и по нескольким уровням индекса:

```python
sensor_data = pd.read_csv('sensor.csv', sep=';')

def print_groups(group_object):
    for name, group in group_object:
        print(name)
        print(group[:5])

# создаём копию данных и заново индексируем её
mi = sensor_data.copy()
mi = mi.set_index(['sensor', 'axis'])

# группируем по первому уровню индекса
print(print_groups(mi.groupby(level=0)))

# группируем по нескольким уровням индекса
print(print_groups(mi.groupby(level=['sensor', 'axis'])))
```

6. Применение агрегирующих функций, преобразований и фильтров. Приведите примеры.

**Ответ**: На этапе применения для каждой группы данных можно выполнить три различные операции:

- Применение агрегирующей функции.
- Преобразование групп данных.
- Фильтрация — исключение всей группы из результатов.

6.1. Применение агрегирующих функций к группам

**Ответ**: Агрегирующие функции можно применить к каждой группе с помощью метода `.aggregate()` (или более короткого `.agg()`) объекта GroupBy. Параметр `.agg()` — это ссылка на функцию, которая будет применяться к каждой группе. В случае с объектом DataFrame эта функция будет применяться к каждому столбцу данных внутри группы.

Кроме того, можно применить несколько агрегирующих функций в рамках одной инструкции, передав список функций. К каждому столбцу можно применить свою функцию, передав в `.agg()` питоновский словарь (ключ — имя столбца, значение — название функции). Агрегирование можно применить к определённым столбцам с помощью оператора `[]` объекта GroupBy.

Для создания числового индекса и переноса уровней исходного индекса в столбцы можно воспользоваться параметром `as_index=False`.

```python
sensor_data = pd.read_csv('sensor.csv', sep=';')

# создаём копию данных и заново индексируем её
mi = sensor_data.copy()
mi = mi.set_index(['sensor', 'axis'])
sensor_axis_grouping = mi.groupby(level=['sensor', 'axis'])

# можно просто применить агрегирующую функцию к группе
print(sensor_axis_grouping.mean())

# применяем сразу несколько агрегирующих функций
print(sensor_axis_grouping.agg([np.sum, np.std]))

# применяем к каждому столбцу свою функцию
print(sensor_axis_grouping.agg({'interval': len, 'reading': np.mean}))

# вычисляем среднее только для столбца reading
print(sensor_axis_grouping['reading'].mean())
```

6.2. Преобразование групп данных

**Ответ**: Объект GroupBy предлагает метод `.transform()`, который применяет ту или иную функцию ко всем значениям объекта DataFrame в каждой группе. Метод `.transform()` объекта GroupBy применяет функцию к каждому значению объекта DataFrame и возвращает другой объект DataFrame, который:

- Идентично проиндексирован, чтобы выполнить конкатенацию индексов во всех группах.
- Имеет количество строк, равное количеству строк, просуммированному по всем группам.
- Состоит из несгруппированных столбцов, к которым библиотека pandas успешно применила заданную функцию (некоторые столбцы могут быть исключены).

Библиотека pandas пытается применить функцию ко всем столбцам, но если столбцы имеют строковые значения, функция терпит неудачу и эти столбцы исключаются из результата. Итоговый объект получит индекс, соответствующий индексу исходного объекта DataFrame.

**Заполнение пропущенных значений групповым средним**

Распространённый вид преобразования — замена пропущенных данных в каждой группе групповым средним (рассчитывается на основе непропущенных значений).

```python
def print_groups(group_object):
    for name, group in group_object:
        print(name)
        print(group[:5])

# создаём данные, чтобы продемонстрировать способ замены пропусков
df = pd.DataFrame({'Label': list("ABABAB"),
                   'Values': [10, 20, 11, np.nan, 12, 22]})
grouped = df.groupby('Label')
print(print_groups(grouped))

# вычисляем среднее для каждой группы
print(grouped.mean())

# заполняем пропущенные значения групповым средним
print(grouped.transform(lambda g: g.fillna(g.mean())))
```

7. Вычисление нормализованных z-значений с помощью преобразования. Приведите примеры.

**Ответ**: Ещё один распространённый вид преобразования — вычисление нормализованных z-значений по группам данных. Для иллюстрации берётся сгенерированный случайным образом ряд значений, используя нормальное распределение со средним значением 0,5 и стандартным отклонением 2. Данные индексируются по дням и на основе 100-дневного окна вычисляется скользящее среднее для каждого значения. Затем скользящие средние стандартизируются по каждому календарному году — данные группируются по годам, задаётся функция, вычисляющая z-значения, и применяется в качестве преобразования к каждой группе.

```python
def print_groups(group_object):
    for name, group in group_object:
        print(name)
        print(group[:5])

# генерируем временной ряд со скользящими средними
np.random.seed(123456)
data = pd.Series(np.random.normal(0.5, 2, 365*3),
                 pd.date_range('2013-01-01', periods=365*3))
periods = 100
rolling = data.rolling(
    window=periods,
    min_periods=periods,
    center=False).mean().dropna()
print(rolling[:5])

# группируем данные по годам и сообщаем о фактическом среднем
# значении и стандартном отклонении каждой группы
print(rolling.groupby(rolling.index.year).mean())
print(rolling.groupby(rolling.index.year).std())

# задаём функцию, вычисляющую z-значения
def zscore(x):
    return (x - x.mean()) / x.std()

# применяем её в качестве преобразования к каждой группе
normed = rolling.groupby(rolling.index.year).transform(zscore)

# сообщаем о новом среднем значении и новом стандартном отклонении
print(normed.groupby(normed.index.year).mean())
print(normed.groupby(normed.index.year).std())
```

8. Исключение групп из процедуры агрегирования. Приведите примеры.

**Ответ**: С помощью метода `.filter()` можно выборочно удалить группы данных из обработки. В этот метод передают функцию, с помощью которой по каждой группе можно принять решение о включении данной группы в итоговый результат. Если функция возвращает значение True, группа включается в результат; если функция возвращает значение False, группа исключается.

Примеры:

1. Удаление групп, которые не соответствуют установленному минимуму элементов (удаляем группы с одним непропущенным значением и меньше).
2. Удаление групп с любым количеством пропущенных значений.
3. Отбор только тех групп, в которых среднее значение превышает значение 2.0 (среднее значение всего набора данных).

```python
# создаём данные для примеров
df = pd.DataFrame({'Label': list('AABCCC'),
                   'Values': [1, 2, 3, 4, np.nan, 8]})
print(df)

# удаляем группы с одним непропущенным значением и меньше
print(df.groupby('Label').filter(lambda x: x.Values.count() > 1))

# удаляем группы с любым количеством пропущенных значений
print(df.groupby('Label').filter(lambda x: x.Values.count() == len(x)))

# отбираем только те группы, в которых среднее значение превышает 2.0
print(df.groupby('Label').filter(lambda x: x.Values.mean() > 2.0))
```
