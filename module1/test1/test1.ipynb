## Вопрос 7

## Вопрос 9

Укажите год, в котором медианное значение стоимости открытия акций компании Microsoft было больше 90$ **Примечание**. С помощью метода `resample()` можно перегруппировать данные по годам.

```python
msft = pd.read_excel('stocks.xlsx', sheet_name='msft')
msft.Date = pd.to_datetime(msft.Date)
msft = msft.set_index('Date')
msft = msft.resample('YE').median()
msft[msft.Open > 90].index.item()
```

## Вопрос 10

Считайте данные с файла. Укажите кол-во дней, когда стоимость закрытия акций компании Microsoft была от 60$ до 80$ включительно.

```python
msft = pd.read_json('stocks.json')
len(msft[
  (msft.Close >= 60)
  &
  (msft.Close <= 80)
])
```
