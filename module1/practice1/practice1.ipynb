# Часть 4. Работа с базой данных SQL

```python
# 1. Используя встроенную Python-библиотеку для работы с базами данных - sqlite3
# создайте базу данных для хранения сведений о работе кафе со следующей структурой:
# [структура]
# Самостоятельно подумайте, какие типы данных следует выбрать для полей таблиц.

import sqlite3
conn = sqlite3.connect("cafe.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS Dishes (
    code INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    price REAL NOT NULL
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS Orders (
    number INTEGER PRIMARY KEY,
    date TEXT NOT NULL
)
""")

cursor.execute("""
CREATE TABLE IF NOT EXISTS Ordered (
    code INTEGER PRIMARY KEY,
    order_number INTEGER,
    dish_code INTEGER,
    FOREIGN KEY(order_number) REFERENCES Orders(number),
    FOREIGN KEY(dish_code) REFERENCES Dishes(code)
)
""")

conn.commit()
```

```python
# 2. Добавьте в таблицу "Блюда" 5 любых блюд, в таблицу "Заказы" - 3 заказа и в каждом заказе по 2-3 блюда (таблица "Заказано").
dishes = [
  (1, "Пицца", 450),
  (2, "Бургер", 320),
  (3, "Паста", 380),
  (4, "Салат", 250),
  (5, "Суп", 200)
]

cursor.executemany("INSERT INTO Dishes VALUES (?, ?, ?)", dishes)

orders = [
  (1, "2025-02-01"),
  (2, "2025-02-02"),
  (3, "2025-02-03")
]

cursor.executemany("INSERT INTO Orders VALUES (?, ?)", orders)

ordered = [
  (1, 1, 1),
  (2, 1, 2),
  (3, 2, 3),
  (4, 2, 4),
  (5, 3, 1),
  (6, 3, 5),
  (7, 3, 4)
]

cursor.executemany("INSERT INTO Ordered VALUES (?, ?, ?)", ordered)
conn.commit()
```

```python
# 3. С помощью метода read_sql выполните извлечение данных из базы данных и представьте в виде датафрейма:
# 3.1. информацию о заказанных блюдах со следующими столбцами: номер заказа, название блюда, цена.

query = """
SELECT Orders.number AS order_number,
       Dishes.name,
       Dishes.price
FROM Ordered
JOIN Orders ON Ordered.order_number = Orders.number
JOIN Dishes ON Ordered.dish_code = Dishes.code
"""

df1 = pd.read_sql(query, conn)
print(df1)
```

```python
# 3.2. сумму к оплате для каждого заказа.

query = """
SELECT Orders.number,
       SUM(Dishes.price) AS total
FROM Ordered
JOIN Orders ON Ordered.order_number = Orders.number
JOIN Dishes ON Ordered.dish_code = Dishes.code
GROUP BY Orders.number
"""

df2 = pd.read_sql(query, conn)
print(df2)
```

```python
# 3.3. названия всех блюд, у которых цена меньше 300 руб.

query = """
SELECT name
FROM Dishes
WHERE price < 300
"""

df3 = pd.read_sql(query, conn)
print(df3)
```

```python
# 3.4. все блюда заказа с номером 2 (или другим номером).

query = """
SELECT Dishes.name
FROM Ordered
JOIN Dishes ON Ordered.dish_code = Dishes.code
WHERE Ordered.order_number = 2
"""

df4 = pd.read_sql(query, conn)
print(df4)
```

```python
# 3.5. список блюд по убыванию цены.

query = """
SELECT name, price
FROM Dishes
ORDER BY price DESC
"""

df5 = pd.read_sql(query, conn)
print(df5)
```

```python
# Измените расчёт суммы к оплате, предусмотрев скидку 5% при наличии в заказе 3 и более блюд.

query = """
SELECT Orders.number,
CASE
    WHEN COUNT(Dishes.code) >= 3
    THEN SUM(Dishes.price) * 0.95
    ELSE SUM(Dishes.price)
END AS total_with_discount
FROM Ordered
JOIN Orders ON Ordered.order_number = Orders.number
JOIN Dishes ON Ordered.dish_code = Dishes.code
GROUP BY Orders.number
"""

df_discount = pd.read_sql(query, conn)
print(df_discount)
```

```python
# 4. Создайте датафреймы для хранения информации о трёх новых блюдах,
# двух новых заказах, в каждом из которых по 2-3 блюда.
# С помощью метода to_sql выполните запись в базу данных.

new_dishes = pd.DataFrame({
    "code": [6,7,8],
    "name": ["Роллы", "Стейк", "Чай"],
    "price": [500, 700, 120]
})

new_dishes.to_sql("Dishes", conn, if_exists="append", index=False)

new_orders = pd.DataFrame({
    "number": [4,5],
    "date": ["2025-02-04", "2025-02-05"]
})

new_orders.to_sql("Orders", conn, if_exists="append", index=False)

new_ordered = pd.DataFrame({
    "code": [8,9,10,11],
    "order_number": [4,4,5,5],
    "dish_code": [6,8,7,3]
})

new_ordered.to_sql("Ordered", conn, if_exists="append", index=False)

# Закрытие соединения
conn.close()
```
