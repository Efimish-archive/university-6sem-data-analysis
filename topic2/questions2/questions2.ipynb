2. Как работать с пропущенными данными? Привести **свои** примеры.

**Ответ**: Данные являются пропущенными (missing), если они содержат значения NaN (также обозначаются как np.nan). Значение NaN означает, что для конкретной метки индекса в конкретной серии не указано значение.

Причины, по которым значение будет записано как значение NaN:
- Объединение двух наборов данных с несовпадающими названиями переменных;
- Данные, извлечённые из внешнего источника, являются неполными;
- Значение переменной неизвестно в данный момент времени и будет заполнено позднее;
- При сборе информации был допущен пропуск, однако наблюдение всё равно должно быть записано;
- Переиндексация данных привела к индексу, у которого отсутствует значение;
- Форма данных изменилась со временем и теперь появились дополнительные строки или столбцы, которые до момента появления этих изменений нельзя было зафиксировать.

Пример создания датафрейма с пропущенными значениями:

```python
# создаём датафрейм с 5 строками и 3 столбцами
df = pd.DataFrame(
  np.arange(0, 15).reshape(5, 3),
  index=['a', 'b', 'c', 'd', 'e'],
  columns=['c1', 'c2', 'c3']
)

# добавляем столбец c4 со значениями NaN
df['c4'] = np.nan
# строка 'f' со значениями от 15 до 18
df.loc['f'] = np.arange(15, 19)
# строка 'g', состоящая из значений NaN
df.loc['g'] = np.nan
# столбец 'c5', состоящий из значений NaN
df['c5'] = np.nan
# меняем значение в столбце 'c4' строки 'a'
df.at['a', 'c4'] = 20
print(df)
```

3. Как найти значения NaN в данных? Привести **свои** примеры.

**Ответ**: Значения NaN в объекте DataFrame можно найти с помощью метода `.isnull()`. Любое значение True означает, что элемент в этой позиции является значением NaN.

Можно использовать тот факт, что метод `.sum()` рассматривает True как 1 и False как 0, чтобы вычислить количество значений NaN в объекте DataFrame. Применение метода `.sum()` к полученной серии даёт общее количество значений NaN в исходном объекте DataFrame.

Ещё один способ определить пропуски — использовать метод `.count()` объекта Series и объекта DataFrame. Для объекта Series этот метод возвращает число непропущенных значений (значений, отличных от NaN). Для объекта DataFrame он будет подсчитывать количество непропущенных значений в каждом столбце.

Кроме того, можно определить, является ли элемент непропущенным значением, воспользовавшись методом `.notnull()`, который возвращает True, если значение является непропущенным, в противном случае он возвращает False.

```python
df = pd.DataFrame(
  np.arange(0, 15).reshape(5, 3),
  index=['a', 'b', 'c', 'd', 'e'],
  columns=['c1', 'c2', 'c3']
)
df['c4'] = np.nan
df.loc['f'] = np.arange(15, 19)
df.loc['g'] = np.nan
df['c5'] = np.nan
df.at['a', 'c4'] = 20

# находим значения NaN с помощью .isnull()
print("isnull():")
print(df.isnull())

# подсчёт NaN по столбцам
print("\nКоличество NaN по столбцам:")
print(df.isnull().sum())

# общее количество NaN
print("\nОбщее количество NaN:", df.isnull().sum().sum())

# количество непропущенных значений по столбцам
print("\n.count() по столбцам:")
print(df.count())

# количество NaN другим способом
print("\nКоличество NaN (через count):")
print(len(df) - df.count())

# проверка непропущенных значений
print("\nnotnull():")
print(df.notnull())
```

4. Как отфильтровывать (удалять) пропущенные данные? Привести **свои** примеры.

**Ответ**: Один из методов обработки пропущенных данных — простое удаление наблюдений с пропусками из вашего набора данных. Библиотека pandas делает это возможным с помощью нескольких методов.

Один из них — логический отбор с использованием результатов методов `.isnull()` или `.notnull()`, которые извлекают из объекта Series пропущенные и непропущенные значения соответственно.

Кроме того, библиотека pandas предлагает удобный метод `.dropna()`, который удаляет пропущенные значения в серии. Обратите внимание, что метод `.dropna()` фактически возвращает копию датафрейма без некоторых строк. Исходный датафрейм не изменился.

Когда метод `.dropna()` применяется к объекту DataFrame, он удаляет из объекта DataFrame все строки, в которых есть хотя бы одно значение NaN.

Если вы хотите удалить только те строки, в которых все значения являются значениями NaN, вы можете использовать параметр `how='all'`.

Кроме того, этот программный код можно применить к столбцам вместо строк, изменив ось строк (используется по умолчанию) на ось столбцов, то есть задав `axis=1`.

Метод `.dropna()` имеет параметр `thresh`. Параметр `thresh` с помощью целочисленного значения задаёт минимальное количество значений NaN, которое требуется, чтобы выполнить операцию удаления строк или столбцов.

Если вы хотите удалить данные в исходном датафрейме, используйте параметр `inplace=True`.

```python
df = pd.DataFrame(
  np.arange(0, 15).reshape(5, 3),
  index=['a', 'b', 'c', 'd', 'e'],
  columns=['c1', 'c2', 'c3']
)
df['c4'] = np.nan
df.loc['f'] = np.arange(15, 19)
df.loc['g'] = np.nan
df['c5'] = np.nan
df.at['a', 'c4'] = 20

# логический отбор непропущенных значений из столбца c4
print("Непропущенные значения столбца c4:")
print(df['c4'][df['c4'].notnull()])

# .dropna() удаляет строки с хотя бы одним NaN
print("\n.dropna() — строки с хотя бы одним NaN удалены:")
print(df.dropna())

# how='all' — удаляем только строки, где ВСЕ значения NaN
print("\n.dropna(how='all') — удалены только строки, где все NaN:")
print(df.dropna(how='all'))

# how='all', axis=1 — удаляем столбцы, где все значения NaN
print("\n.dropna(how='all', axis=1) — удалён столбец c5:")
print(df.dropna(how='all', axis=1))

# thresh=5, axis=1 — удаляем столбцы с не менее чем 5 значениями NaN
print("\n.dropna(thresh=5, axis=1) — удалены столбцы c4 и c5:")
print(df.dropna(thresh=5, axis=1))
```

5. Как pandas обрабатывает пропущенные значения в ходе вычислений? Привести **свои** примеры.

**Ответ**: Значения NaN в библиотеке pandas обрабатываются иначе, чем в библиотеке NumPy. Когда функция библиотеки NumPy встречает значение NaN, она возвращает NaN. Функции библиотеки pandas обычно игнорируют значения NaN и продолжают выполнять обработку данных, как если бы значения NaN не были частью объекта Series.

Способ, с помощью которого библиотека pandas обрабатывает значения NaN:
- При суммировании данных значения NaN обрабатываются как нулевые значения;
- Если все значения являются значениями NaN, результатом будет значение NaN;
- Методы типа `.cumsum()` и `.cumprod()` игнорируют значения NaN, но сохраняют их в полученных массивах.

Но при использовании традиционных математических операторов значения NaN будут просто перенесены в итоговую серию.

```python
df = pd.DataFrame(
  np.arange(0, 15).reshape(5, 3),
  index=['a', 'b', 'c', 'd', 'e'],
  columns=['c1', 'c2', 'c3']
)
df['c4'] = np.nan
df.loc['f'] = np.arange(15, 19)
df.loc['g'] = np.nan
df['c5'] = np.nan
df.at['a', 'c4'] = 20

s = df.c4

# значения NaN обрабатываются как 0 при суммировании
print("sum() столбца c4:", s.sum())

# среднее вычисляется только по непропущенным значениям
print("mean() столбца c4:", s.mean())

# cumsum() игнорирует NaN, но сохраняет их в результате
print("\ncumsum() столбца c4:")
print(s.cumsum())

# при арифметических операциях NaN переносится в результат
print("\ndf.c4 + 1:")
print(df.c4 + 1)
```

6. Как найти, отфильтровать и исправить неизвестные значения? Привести **свои** примеры.

**Ответ**: Вместо того, чтобы оставить пропуски как есть или удалить строки/столбцы, содержащие их, можно воспользоваться методом `.fillna()`, который заменяет значения NaN на определённое значение.

**Прямое и обратное заполнение.** Пропуски в данных можно заполнить с помощью последнего непропущенного значения в прямом и обратном порядке. При работе с данными, представленными в виде временных рядов, этот метод заполнения часто называют методом «последнего известного значения». Для прямого заполнения используется `method='ffill'`, для обратного — `method='bfill'`. Библиотека pandas также предлагает глобальные функции `pd.ffill()` и `pd.bfill()`.

**Заполнение с помощью меток индекса.** Данные можно заполнить с помощью меток объекта Series или ключей питоновского словаря. Это позволяет задавать для заполнения разные значения для различных элементов, используя индексные метки. Заполняются только значения NaN.

**Заполнение значений NaN средним значением столбца.** Ещё один распространённый сценарий — заполнение всех значений NaN в столбце средним значением столбца. Это удобно, поскольку пропуски, замененные таким способом, искажают статистическое среднее в меньшей степени, чем заполнение пропусков нулями.

```python
df = pd.DataFrame(
  np.arange(0, 15).reshape(5, 3),
  index=['a', 'b', 'c', 'd', 'e'],
  columns=['c1', 'c2', 'c3']
)
df['c4'] = np.nan
df.loc['f'] = np.arange(15, 19)
df.loc['g'] = np.nan
df['c5'] = np.nan
df.at['a', 'c4'] = 20

# 1. Замена NaN на определённое значение (0)
print("fillna(0):")
filled = df.fillna(0)
print(filled)

# Примечание: возникла проблема из-за изменения API в pandas ≥ 2.2:
# параметр method= был удалён из .fillna().
# Вместо него теперь нужно использовать самостоятельные функции pd.ffill() и pd.bfill().

# 2. Прямое заполнение (ffill)
print("\nПрямое заполнение (ffill) столбца c4:")
# print(df.c4.fillna(method='ffill'))
print(df.c4.ffill())

# 3. Обратное заполнение (bfill)
print("\nОбратное заполнение (bfill) столбца c4:")
# print(df.c4.fillna(method='bfill'))
print(df.c4.bfill())

# 4. Заполнение с помощью меток индекса
fill_values = pd.Series([100, 101, 102], index=['a', 'e', 'g'])
print("\nЗаполнение с помощью меток индекса (столбец c4):")
print(df.c4.fillna(fill_values))

# 5. Заполнение средним значением столбца
print("\nЗаполнение средним значением столбца:")
print(df.fillna(df.mean()))
```

7. Выполнение интерполяции пропущенных значений. Привести **свои** примеры.

**Ответ**: Метод `.interpolate()` объекта Series и объекта DataFrame выполняет по умолчанию линейную интерполяцию пропущенных значений.

Значение, используемое для интерполяции, вычисляется следующим образом. Сначала на основе первого значения, предшествующего ряду значений NaN, и последнего значения, следующего после ряда значений NaN, вычисляем инкремент (величину приращения). Затем пошагово увеличиваем стартовое значение на величину этого инкремента и заполняем им значения NaN.

Кроме того, метод интерполяции позволяет указать конкретный метод интерполяции. Одним из распространённых методов является интерполяция на основе времени — `method='time'`.

Помимо этого интерполяцию можно задать для вычисления значений относительно меток индекса, когда используются числовые индексные метки, — `method='values'`.

```python
# линейная интерполяция значений NaN с 1 по 2
s = pd.Series([1, np.nan, np.nan, np.nan, 2])
print("Линейная интерполяция:")
print(s.interpolate())

# интерполяция на основе индексных меток (method='values')
s2 = pd.Series([0, np.nan, 100], index=[0, 1, 10])
print("\nЛинейная интерполяция (без учёта меток индекса):")
print(s2.interpolate())

print("\nИнтерполяция относительно меток индекса (method='values'):")
print(s2.interpolate(method='values'))
```

8. Как найти и удалить дублирующиеся наблюдения? Привести **свои** примеры.

**Ответ**: Дублирующиеся данные считаются ошибкой, особенно в тех случаях, когда эти данные можно считать идемпотентными.

Библиотека pandas предлагает метод `.duplicated()` для упрощения поиска дублирующихся данных. Этот метод возвращает серию логических значений, в которой каждая запись показывает, является ли строка дубликатом или нет. Значение True означает, что данная строка уже появлялась ранее в объекте DataFrame, значения в столбцах совпадают.

Дублирующиеся строки можно удалить из датафрейма с помощью метода `.drop_duplicates()`. Этот метод возвращает копию датафрейма с удалёнными дублирующимися строками. Кроме того, можно воспользоваться параметром `inplace=True`, чтобы удалить строки, не создавая копию датафрейма.

По умолчанию при удалении дубликатов каждый раз оставляется строка, встретившаяся первой. Если вы хотите, чтобы при удалении дубликатов каждый раз оставлялась последняя строка, используйте параметр `keep='last'`.

Если вы хотите проверить наличие дубликатов, используя меньший набор столбцов, вы можете задать список имён столбцов.

```python
data = pd.DataFrame({
  'a': ['x', 'x', 'x', 'y', 'y', 'y', 'y'],
  'b': [1, 1, 2, 3, 3, 4, 4]
})
print("Исходный датафрейм:")
print(data)

# находим дублирующиеся строки
print("\n.duplicated():")
print(data.duplicated())

# удаляем дублирующиеся строки (оставляем первое вхождение)
print("\n.drop_duplicates():")
print(data.drop_duplicates())

# оставляем последнее вхождение
print("\n.drop_duplicates(keep='last'):")
print(data.drop_duplicates(keep='last'))

# добавляем столбец c и проверяем дубликаты по столбцам a и b
data['c'] = range(7)
print("\n.duplicated() после добавления столбца c (все False):")
print(data.duplicated())

print("\n.drop_duplicates(['a', 'b']) — дублирование проверяется только по a и b:")
print(data.drop_duplicates(['a', 'b']))
```

9. Как преобразовать значения с помощью методов .replace(), .map() и apply()? Привести **свои** примеры.

**Ответ**:

**Метод `.map()`** — одна из основных задач преобразования данных — это сопоставление набора значений другому набору значений. Библиотека pandas позволяет сопоставлять значения с помощью таблицы поиска (по питоновскому словарю или серии библиотеки pandas). Метод `.map()` выполняет сопоставление, сперва сличая значения внешней серии с метками индекса внутренней серии. Затем он возвращает новую серию с метками индекса внешней серии, но со значениями внутренней серии. Если библиотека pandas не найдёт соответствия между значением внешней серии и индексной меткой внутренней серии, она пометит это значение как пропущенное.

**Метод `.replace()`** — метод `.fillna()` можно рассматривать как специальный случай более универсальной операции замены значений, которая реализована в рамках метода `.replace()`. Этот метод более гибок благодаря возможности заменить любое значение (не только значение NaN) другим значением. Можно задать несколько элементов для замены, передав два списка или словарь. Если используется метод `.replace()` объекта DataFrame, можно указать разные заменяемые значения для каждого столбца.

**Метод `.apply()`** — при работе с данными можно применить различные функции, воспользовавшись методом `.apply()`. Когда задана функция Python, метод `.apply()` итеративно вызывает её, передавая каждое значение из объекта Series. Метод `.apply()` объекта DataFrame передаёт в функцию каждый столбец, представленный в виде объекта Series, или, если задан параметр `axis=1`, передаёт объект Series, представляющий каждую строку.

**Метод `.applymap()`** — в то время как метод `.apply()` всегда применяет функцию к строке или столбцу целиком, метод `.applymap()` применяет функцию к каждому отдельному значению.

```python
# ── .replace() ──
s = pd.Series([0., 1., 2., 3., 2., 4.])

# замена одного значения
print(".replace(2, 5):")
print(s.replace(2, 5))

# замена нескольких значений через два списка
print("\n.replace([0,1,2,3,4], [4,3,2,1,0]):")
print(s.replace([0, 1, 2, 3, 4], [4, 3, 2, 1, 0]))

# замена через словарь
print("\n.replace({0: 10, 1: 100}):")
print(s.replace({0: 10, 1: 100}))

# замена в DataFrame — разные значения для каждого столбца
df = pd.DataFrame({'a': [0, 1, 2, 3, 4], 'b': [5, 6, 7, 8, 9]})
print("\n.replace({'a': 1, 'b': 8}, 100) в DataFrame:")
print(df.replace({'a': 1, 'b': 8}, 100))

# ── .map() ──
x = pd.Series({'1': 'a', '2': 'b', '3': 'c'})
y = pd.Series({'a': 1, 'b': 2, 'c': 3})
print("\n.map() — сопоставление значений:")
print(x.map(y))

# ── .apply() к Series ──
s2 = pd.Series([1, 2, 3, 4])
print("\n.apply(lambda x: x * 2) к Series:")
print(s2.apply(lambda x: x * 2))

# ── .apply() к DataFrame (по столбцам и строкам) ──
df2 = pd.DataFrame(np.arange(12).reshape(4, 3), columns=['a', 'b', 'c'])
print("\n.apply(sum) по столбцам:")
print(df2.apply(lambda col: col.sum()))

print("\n.apply(sum) по строкам (axis=1):")
print(df2.apply(lambda row: row.sum(), axis=1))

# создание нового столбца через .apply()
df2['interim'] = df2.apply(lambda r: r.a * r.b, axis=1)
df2['result'] = df2.apply(lambda r: r.interim + r.c, axis=1)
print("\nDataFrame с новыми столбцами interim и result:")
print(df2)

# ── .map() ──
df3 = pd.DataFrame(np.arange(0, 15).reshape(3, 5))
df3.loc[1, 2] = np.nan
print("\n.map(lambda x: '%.2f' % x):")
print(df3.map(lambda x: '%.2f' % x))
```
