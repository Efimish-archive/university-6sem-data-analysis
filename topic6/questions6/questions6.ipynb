6. Что такое генеральная совокупность? Привести пример.

**Ответ**: **Генеральная совокупность** — это совокупность всех объектов, которые представляют интерес в конкретном исследовании.

Примеры из лекции:
- Если вы хотите узнать, как женщины старше 18 в России относятся к новой рекламе какого-либо продукта, то генеральной совокупностью будут абсолютно все женщины, проживающие в России, чей возраст превышает 18 лет.
- Если вам интересен прогресс в обучении у первоклассников, то генеральной совокупностью будут абсолютно все дети, которые обучаются в школе первый год.

7. Что такое выборка? Привести пример.

**Ответ**: **Выборка** — это группа объектов, отобранных из генеральной совокупности для исследования.

Из генеральной совокупности отбирается только часть объектов (выборка), проводится на них исследование, а получившиеся результаты обобщаются на всю генеральную совокупность.

Пример из лекции: есть 1000 школьников, которые сдали ЕГЭ — это генеральная совокупность. Из этой группы выбираются две выборки по 5 человек.

8. Репрезентативность выборки. Смещение.

**Ответ**: **Репрезентативность** — соответствие характеристик выборки характеристикам генеральной совокупности.

Для того чтобы было выполнено условие репрезентативности, все основные особенности исследуемой группы людей или объектов должны быть одни и те же у выборки и у генеральной совокупности. Например, если в генеральной совокупности 40% женщин и 60% мужчин, то и в выборке должно быть такое же соотношение. При соблюдении процедуры такую выборку можно, например, получить с помощью случайного выбора.

Важно понимать, что даже для репрезентативной выборки будет наблюдаться **ошибка выборки** — несоответствие между характеристиками генеральной совокупности и характеристиками, полученными для данной выборки.

**Смещение выборки** — явление, при котором статистические характеристики выборки сильно отличаются (смещены) относительно характеристик генеральной совокупности.

Причины смещения (из лекции):
1. Нерепрезентативная выборка (например, психолог изучает особенности восприятия людей в целом, но в качестве респондентов использует только своих студентов).
2. Отсутствие части данных (в социальных опросах люди могут избегать ответов на вопросы про религию, деньги и другие личные темы).
3. Явление социальной желательности (люди не склонны отвечать даже анонимно то, что не вызовет социальное одобрение).

10. Что такое случайная величина (СВ)? Привести пример дискретной и непрерывной СВ.

**Ответ**: Под **случайной величиной** понимается переменная, которая в результате испытания в зависимости от случая принимает одно из возможного множества своих значений.

**Дискретная случайная величина** — случайная величина, множество значений которой конечное, или бесконечное, но счётное.

Пример дискретной СВ (из лекции): число бракованных изделий на линии; результаты футбольного матча; температура тела человека.

**Непрерывная случайная величина** — случайная величина, которая может принимать любые значения из некоторого промежутка.

Пример непрерывной СВ (из лекции): цена доллара; убытки при наступлении страхового случая.

11. Что такое закон распределения СВ?

**Ответ**: **Законом распределения случайной величины** называется всякое соотношение, устанавливающее связь между возможными значениями случайной величины и соответствующими им вероятностями.

Закон распределения для дискретной случайной величины может быть задан в виде ряда распределения случайной величины — таблицы, в первой строке которой в порядке возрастания указаны возможные значения случайной величины, а во второй строке — соответствующие вероятности. При этом p_i = P(X = x_i) — вероятность события (X = x_i). Все события (X = x_i) образуют полную группу событий, поэтому сумма всех p_i равна 1.

Ряд распределения может быть изображён графически: по оси абсцисс откладывают значения случайной величины, а по оси ординат — соответствующие им вероятности. Соединение полученных точек образует ломаную, называемую многоугольником или полигоном распределения вероятности.

12. Числовые характеристики дискретных СВ.

**Ответ**: **Математическое ожидание** дискретной случайной величины — сумма произведений всех её возможных значений на соответствующие им вероятности:

M(X) = Σ x_i · p_i = x₁p₁ + x₂p₂ + … + xₙpₙ

Свойства математического ожидания:
1. M(C) = C, C — const.
2. M(kX) = k·M(X).
3. M(X ± Y) = M(X) + M(Y).
4. M(X · Y) = M(X) · M(Y), если X, Y — независимы.
5. M(X − M(X)) = 0.

**Дисперсия** случайной величины — математическое ожидание квадрата её отклонений от математического ожидания:

D(X) = M(X − M(X))²

Дисперсия может быть рассчитана по формуле:

D(X) = M(X²) − (M(X))²

Свойства дисперсии:
1. D(C) = 0, C — const.
2. D(kX) = k²·D(X).
3. D(X ± Y) = D(X) + D(Y), если X, Y — независимы.

**Среднее квадратическое отклонение (стандартное отклонение)** случайной величины — корень квадратный из её дисперсии:

σ(X) = √D(X)

**Мода** дискретной случайной величины — возможное значение случайной величины, которое имеет наибольшую соответствующую вероятность.

**Медиана** дискретной случайной величины — возможное значение случайной величины, слева и справа от которого в ряде распределения случайной величины одинаковое число значений случайной величины.

13. Функция распределения СВ, свойства.

**Ответ**: **Функцией распределения** случайной величины X называется функция F(x), выражающая для каждого события x вероятность того, что случайная величина X примет значение, меньшее x:

F(x) = P(X < x)

**Свойства функции распределения:**
1. 0 ≤ F(x) ≤ 1.
2. F(x) — монотонно не убывает: если x₂ > x₁, то F(x₂) ≥ F(x₁).
3. lim F(x) при x→−∞ = 0; lim F(x) при x→+∞ = 1.
4. P(x₁ ≤ X < x₂) = F(x₂) − F(x₁).

Случайная величина X называется непрерывной, если её функция распределения непрерывна в любой точке и дифференцируема всюду, кроме, может быть, отдельных точек.

Вероятность любого отдельно взятого значения непрерывной случайной величины равна нулю: P(X = x) = 0.

Для непрерывной случайной величины справедливо равенство:

P(x₁ ≤ X ≤ x₂) = P(x₁ ≤ X < x₂) = P(x₁ < X ≤ x₂) = P(x₁ < X < x₂)

14. Функция плотности вероятности непрерывной СВ, свойства.

**Ответ**: **Плотностью вероятности (дифференциальной функцией распределения)** непрерывной случайной величины называется производная её функции распределения:

φ(x) = F′(x)

Для непрерывной случайной величины справедливо равенство:

P(a ≤ X ≤ b) = ∫ от a до b φ(x) dx

Функция распределения непрерывной случайной величины может быть выражена через плотность вероятности по формуле:

F(x) = ∫ от −∞ до x φ(t) dt

**Свойства дифференциальной функции распределения:**
1. φ(x) ≥ 0.
2. ∫ от −∞ до +∞ φ(x) dx = 1.

15. Числовые характеристики непрерывных случайных величин.

**Ответ**: **Математическое ожидание:**

M(X) = ∫ от −∞ до +∞ x · φ(x) dx

**Дисперсия:**

D(X) = ∫ от −∞ до +∞ (x − M(X))² · φ(x) dx

16. Смоделировать 5 выборок объемом 1000 элементов из разных законов распределения: нормальное, экспоненциальное, равномерное и два любых других закона распределения на выбор (см. Генераторы распределений). Визуализировать полученные выборки, используя гистограммы и диаграммы «Ящик с усами», построить эмпирические функцию и плотность распределения. Для каждой выборки посчитать описательные характеристики (минимум и максимум, мат. ожидание, дисперсия, стандартное отклонение, квантили уровней 0.25 и 0.75, размах, медиана, коэффициенты асимметрии и эксцесса).

```python
np.random.seed(69)
size = 1000

# Генерация выборок
samples = {
  "Нормальное распределение": np.random.normal(loc=0, scale=1, size=size),
  "Экспоненциальное распределение": np.random.exponential(scale=1, size=size),
  "Равномерное распределение": np.random.uniform(low=0, high=1, size=size),
  "Распределение Пуассона": np.random.poisson(lam=5, size=size),
  "Распределение Хи-квадрат": np.random.chisquare(df=2, size=size),
}

# Визуализация выборок c помощью гистограмм
plt.figure(figsize=(15, 10))

for i, (name, sample) in enumerate(samples.items(), 1):
  plt.subplot(3, 2, i)
  sns.histplot(sample, kde=True)
  plt.ylabel("")
  plt.title(name)

plt.tight_layout()
plt.show()
```

```python
# Визуализация выборок c помощью диаграммы "Ящик с усами"
plt.figure(figsize=(10, 6))

sns.boxplot(list(samples.values()))

plt.xticks(
  ticks=range(len(samples)),
  labels=["Нормальное", "Экспоненциальное", "Равномерное", "Пуассон", "Хи-квадрат"],
)
plt.title("Диаграммы «Ящик с усами» для всех выборок")
plt.show()
```

```python
# Построение эмпирических функций распределения и плотности
plt.figure(figsize=(15, 10))

for i, (name, sample) in enumerate(samples.items(), 1):
  plt.subplot(3, 2, i)
  sns.ecdfplot(sample)
  plt.ylabel("")
  plt.title(name)

plt.tight_layout()
plt.show()
```
