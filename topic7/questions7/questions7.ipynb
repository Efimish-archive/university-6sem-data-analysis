2. Cреднее арифметическое (определение). Как определить  среднее арифметическое генеральной совокупности? Как определить  среднее арифметическое выборки?

**Ответ**:
**Определение:** Среднее арифметическое — число, равное сумме всех чисел выборки или генеральной совокупности, деленной на их количество.

Чтобы рассчитать среднее арифметическое, нужно сложить все значения переменной и разделить эту сумму на их количество. Среднее генеральной совокупности обычно обозначается греческой буквой μ («мю»), а среднее выборки — заглавной латинской буквой M или x̄ («икс с чертой»).

**Формула для среднего арифметического генеральной совокупности:**

μ = (Σ X) / N

**Формула для среднего арифметического выборки:**

M = (Σ X) / n

Σ X здесь значит, что нужно найти сумму всех значений X.

Греческие буквы используются, чтобы обозначить параметры генеральной совокупности, а заглавные латинские буквы для статистик выборок. Заглавная N обозначает количество наблюдений в генеральной совокупности, а строчная n — в выборке.

3. Свойства среднего арифметического и применение.

**Ответ**:
**Балансировочная точка**

Одно из важных свойств среднего арифметического, что оно служит своеобразной балансировочной точкой для всех данных. Общие расстояния до точек, которые находятся выше среднего арифметического и тех, которые находятся ниже него, будут совпадать. Также из этого свойства очевидно, что значение среднего арифметического всегда находится между наименьшим и наибольшим значением нашего распределения — оно не может выходить за пределы этого интервала. Если вдруг вы рассчитали среднее арифметическое и обнаружили, что это не так, то нужно искать ошибку в вычислениях.

**Чувствительность к аномалиям**

Среднее арифметическое очень чувствительно к аномально маленьким или аномально большим значениям. Такие значения еще называются выбросами. Если мы встретим такие в наших данных, то они могут повлиять на нашу меру центральной тенденции и «утянуть» среднее арифметическое за собой вправо или влево.

**Когда использовать среднее арифметическое?**

- Для непрерывных количественных переменных, которые мы хотя бы в теории можем выразить дробными числами.
- Для симметричных распределений.
- В теории, среднее арифметическое можно рассчитать для бинарной переменной, записанной единицами и нулями. Но смысл у него будет другой — такая мера по сути будет отражать долю единиц среди значений переменной.

4. Измерение центральной тенденции (среднего, медианы и моды)

**Ответ**:
Самых популярных мер центральных тенденций три — среднее арифметическое, медиана и мода.

---

**Среднее арифметическое**

Чтобы рассчитать среднее арифметическое, нужно сложить все значения переменной и разделить эту сумму на их количество.

Среднее значение можно вычислить с помощью метода `.mean()`. Библиотека pandas берет каждый столбец и отдельно вычисляет для него среднее значение. Она возвращает результаты в виде значений серии, проиндексированной по именам столбцов. По умолчанию метод `.mean()` применяется к оси строк (axis=0), то есть двигается сверху вниз по каждому столбцу, чтобы вычислить среднее значение для него. Для переключения оси строк на ось столбцов следует задать параметр axis=1, чтобы найти среднее значение для каждой строки двигаясь слева направо по всем столбцам.

---

**Медиана**

**Определение:** Медиана — это точка ниже и выше которой находится ровно по 50% значений в распределении.

У нас нет математического обозначения медианы. Для выборки и генеральной совокупности мы будем рассчитывать медиану одинаково. При необходимости мы будем обозначать ее как median.

Чтобы найти медиану, нам нужно упорядочить значения переменной от меньшего к большему и найти то значение, которое делит распределение пополам. Если в выборке чётное количество элементов — одна из самых популярных конвенций нахождения медианы для распределения с чётным количеством элементов — это среднее арифметическое центральных элементов.

Медиана вычисляется с помощью метода `.median()`.

---

**Мода**

**Определение:** Мода — это значение переменной с самой большой частотой, т.е. самое популярное значение переменной.

Как и с медианой, у нас нет специальных символов для обозначения моды. И для генеральной совокупности, и для выборки мода находится одинаково. При необходимости мы будем обозначать ее как mode.

Чтобы найти моду, нам нужно построить частотную таблицу и определить самое часто встречающееся значение.

Мода вычисляется с помощью метода `.mode()`. Данная строка возвращает не скалярное значение, представляющее моду, а серию. Данный факт связан с тем, что мод может быть несколько.

```python
np.random.seed(69)

values = pd.Series(np.random.randint(10, 20, 20))

print("Среднее арифметическое:", round(values.mean(), 2))
print("Медиана:", round(values.median()))
print("Мода:", ", ".join(map(str, values.mode())))
```

5. Меры вариативности и их назначение. Вычисление размаха, интерквартильного размаха, дисперсии, стандартного отклонения.

**Ответ**:
**Меры вариативности (разброса)** представляют собой количественные меры, выражающие различия между значениями в распределении. Более формально — это степень индивидуальных отклонений значений от центральной тенденции.

Зная меру вариативности, можно представить форму нашего распределения. Вариативность также помогает нам понять, насколько типичной для распределения является случайно взятая оценка.

---

**Размах**

**Определение:** Размах — это разница между самым большим и самым маленьким значением переменной.

Размах не имеет специального обозначения и рассчитывается одинаково и для выборки, и для генеральной совокупности. При необходимости мы будем обозначать размах в формулах словом range.

$$range = X_{max} - X_{min}$$

---

**Интерквартильный размах**

Квартили — это значения, которые делят распределение на четверти.

- Первый квартиль (нижний) — отделяет первые 25% значений от следующих 75%.
- Второй квартиль (он же медиана) — делит выборку пополам (50% и 50%).
- Третий квартиль (верхний) — отделяет первые 75% выборки от следующих 25%.

**Определение:** Интерквартильный размах — интервал значений признака, содержащий центральные 50% наблюдений распределения, то есть интервал между первым и третьим квартилем.

Интерквартильный размах тоже не имеет специального обозначения и рассчитывается одинаково для выборки и генеральной совокупности. Мы будем обозначать его как IQR (inter-quartile range).

$$IQR = Q_3 - Q_1$$

---

**Дисперсия**

**Определение:** Дисперсия — среднее квадратов отклонений от среднего арифметического распределения.

Формула дисперсии генеральной совокупности:

$$\sigma^2 = \frac{\sum(X - \mu)^2}{N}$$

Формула дисперсии выборки:

$$s^2 = \frac{\sum(X - M)^2}{n - 1}$$

В формуле для выборки в знаменателе используется n - 1, а не n — чтобы скорректировать недооценку вариативности генеральной совокупности.

---

**Среднеквадратичное (стандартное) отклонение**

**Определение:** Среднеквадратичное отклонение (стандартное отклонение) — квадратный корень дисперсии. Мера, которая определяет среднее отклонение от среднего арифметического распределения.

Для генеральной совокупности обозначается σ (сигма), для выборки — s. Дисперсия — среднеквадратичное отклонение в квадрате — обозначается σ² и s² соответственно.

---

**Вычисление в Python**

В библиотеке pandas дисперсия вычисляется с помощью метода `.var()`. Стандартное отклонение вычисляется с использованием метода `.std()`. По умолчанию методы `var()` и `std()` вычисляют исправленную выборочную дисперсию и стандартное отклонение соответственно. Если же мы работаем с генеральной совокупностью, то мы можем вычислить не оценку дисперсии по выборочным данным, а истинное значение. Для этого следует указать параметр `ddof=0` (степени свободы).

Квантили можно вычислить непосредственно в pandas при помощи метода `quantile`. Последовательность требующихся квантилей передается в виде списка.

```python
np.random.seed(69)

values = pd.Series(np.random.randint(10, 20, 20))

print("Размах:", values.max() - values.min())
print(
  "Интерквартильный размах (IQR):",
  values.quantile(0.75) - values.quantile(0.25)
)
print("Дисперсия (выборка):", round(values.var(), 2))
print(
  "Дисперсия (генеральная совокупность, ddof=0):",
  round(values.var(ddof=0), 2)
)
print("Стандартное отклонение (выборка):", round(values.std(), 2))
print(
  "Стандартное отклонение (генеральная совокупность):",
  round(values.std(ddof=0), 2)
)
```

6. z-оценка (определение и  назначение). Измерение z- оценки для генеральной совокупности и для выборки. Выражение среднего и среднеквадратичного отклонения через z-оценку.

**Ответ**:
**Определение:** Стандартизированная оценка (z-оценка) — это относительная мера, которая показывает, на сколько среднеквадратичных отклонений наблюдаемое значение отличается от среднего значения распределения. Знак z-оценки показывает, находится ли значение левее среднего (–) или правее среднего (+).

**Назначение:** С помощью z-оценок мы можем стандартизировать распределение — выразить все значения в терминах дистанции от среднего. Это способ выразить любое значение переменной в количестве среднеквадратичных отклонений от среднего распределения.

---

**Формула z-оценки для генеральной совокупности:**

$$z = \frac{X - \mu}{\sigma}$$

**Формула z-оценки для выборки:**

$$z = \frac{X - M}{s}$$

В числителе у нас находится отклонение значения от среднего. А разделив его на среднеквадратичное отклонение, мы находим расстояние от значения до среднего в единицах среднеквадратичных отклонений.

---

**Выражение среднего и среднеквадратичного отклонения через z-оценку**

Зная z-оценку, среднее и среднеквадратичное отклонение, можно легко восстановить и оригинальное значение.

Для генеральной совокупности:

$$X = \mu + z \cdot \sigma$$

Для выборки:

$$X = M + z \cdot s$$

7. Стандартизированное распределение и его свойства.

**Ответ**: Если мы превратим все значения переменной в распределении в z-оценки, то мы получим z-распределение или стандартизированное распределение. Такое распределение сохранит форму оригинального распределения, но среднее в нём будет равно нулю, а среднеквадратичное отклонение — единице.

Z-распределение обладает важными свойствами. Через z-оценки мы можем описать пропорции нормального распределения. Так в интервале от минус одного до плюс одного среднеквадратичного отклонения располагается примерно 68% выборки — то есть самые типичные для неё значения. Знание этих пропорций пригодится и при разговоре о тестировании гипотез, когда мы будем оценивать вероятность случайного получения некоторых значений.

Стандартизированное распределение также позволяет сравнивать значения из распределений с разными шкалами. Например, после того как мы найдём z-оценку для каждого оригинального значения переменной в двух группах, мы сможем сравнить оценки между собой.

8. Выбросы (определение и  назначение).

**Ответ**:
**Определение:** Выброс — это аномальное значение в данных, которое значительно отличается от значения, выраженного мерой центральной тенденции.

Определение выбросов обычно субъективно, но есть некоторые конвенции. Например, значения отличающиеся от среднего на 3 и больше среднеквадратичных отклонений называют сильными выбросами, а от 2 до 3 отклонений – умеренными.

Выбросы можно также выразить через квартили. В популярном виде визуализации «ящик-с-усами» обычно отображаются и выбросы — все что вышло за границы «усов»:

- нижняя граница = Q1 – 1.5 * IQR
- верхняя граница = Q3 + 1.5 * IQR

При этом ограничитель уса находится не в конце рассчитанного отрезка, а на последнем значении набора, попавшем в этот интервал. Оставшиеся точки, выходящие за пределы усов, отмечают на графике как выбросы.

**Зачем находить выбросы?**

Во-первых, так как эти значения аномальны, они могут нести в себе какую-то интересную информацию. Например, в примере с зарплатами имеет смысл посмотреть, а какими такими квалификациями обладает «нетипичный» выпускник.

Во-вторых, как мы уже видели, выбросы могут повлиять на расчеты некоторых метрик. Если значения выбросов действительно нетипичны и обусловлены какими-то понятными причинами нерелевантными вашему исследовательскому вопросу, то от таких наблюдений можно избавиться.

9. Исключение выбросов с помощью Python.

**Ответ**: Одним из способов нахождения выбросов является использование интерквартильного размаха: будем считать выбросами все, что находится ниже 1-й квартили на 1,5 интерквартильных размаха и выше 3-й квартили на 1,5 интерквартильных размаха.

```python
np.random.seed(69)

views = np.random.normal(loc=50, scale=10, size=100)
views = pd.Series(np.append(views, [10, 90]))

q1 = views.quantile(0.25)
q3 = views.quantile(0.75)
iqr = q3 - q1

lower_iqr = q1 - 1.5 * iqr
upper_iqr = q3 + 1.5 * iqr

filtered_iqr = views[views.between(lower_iqr, upper_iqr)]

print(
  "Диапазон без выбросов (по IQR): от",
  round(filtered_iqr.min(), 1),
  "до",
  round(filtered_iqr.max(), 1)
)
```

Ещё одним способом нахождения выбросов является использование среднеквадратического отклонения: чаще всего выбросами считаются наблюдения, находящиеся на расстоянии 3 и более среднеквадратических отклонения от среднего (правило 3-х сигм).

```python
mean = views.mean()
std = views.std()

lower_std = mean - 3 * std
upper_std = mean + 3 * std

filtered_std = views[views.between(lower_std, upper_std)]

print(
  "Диапазон без выбросов (по 3 std): от",
  round(filtered_std.min(), 1),
  "до",
  round(filtered_std.max(), 1)
)
```
